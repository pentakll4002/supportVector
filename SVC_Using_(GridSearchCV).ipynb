# Support Vector Classifier Implementation

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
## Let create synthelic data point
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=2,
                           n_redundant=0)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)[0]
```

```python
pd.DataFrame(y)[0]
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=y)
```

```python
### Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='linear')
```

```python
svc.fit(X_train, y_train)
```

```python
### Prediction
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
acc = accuracy_score(y_test, y_pred)
print(acc)
cm = confusion_matrix(y_test, y_pred)
print(cm)
cr = classification_report(y_test, y_pred)
print(cr)
```

```python
print(svc.intercept_)
print(svc.coef_)
```

# RBF

```python
rbf = SVC(kernel='rbf')
rbf.fit(X_train, y_train)
### Predict
y_pred1 = rbf.predict(X_test)
acc = accuracy_score(y_test, y_pred1)
print(acc)
cm = confusion_matrix(y_test, y_pred1)
print(cm)
cr = classification_report(y_test, y_pred1)
print(cr)
```

```python
print(rbf.intercept_)
```

# Polynomial

```python
polynomial = SVC(kernel='poly')
polynomial.fit(X_train, y_train)
## Predict
y_pred2 = polynomial.predict(X_test)
acc = accuracy_score(y_test, y_pred2)
print(acc)
cm = confusion_matrix(y_test, y_pred2)
print(cm)
cr = classification_report(y_test, y_pred2)
print(cr)
```

```python
print(polynomial.intercept_)
```

#Sigmoid

```python
sigmoid = SVC(kernel='sigmoid')
sigmoid.fit(X_train, y_train)
## Predict
y_pred3 = sigmoid.predict(X_test)
acc = accuracy_score(y_test, y_pred3)
print(acc)
cm = confusion_matrix(y_test, y_pred3)
print(cm)
cr = classification_report(y_test, y_pred3)
print(cr)
```

```python
print(sigmoid.intercept_)
```

# Hyperparameter Turning with SVC (GridSearchCV)


```python
from sklearn.model_selection import GridSearchCV

# defining parameter range
param_grid = {
    'C': [0.1, 1, 10, 100, 1000],
    'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
    'kernel': ['rbf', 'linear', 'polynomial']
}
```

```python
grid = GridSearchCV(estimator=SVC(), param_grid=param_grid, cv=5, refit=True, verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
## Predict
y_pred4 = grid.predict(X_test)
acc = accuracy_score(y_test, y_pred4)
print(acc)
cm = confusion_matrix(y_test, y_pred4)
print(cm)
cr = classification_report(y_test, y_pred4)
print(cr)
```
